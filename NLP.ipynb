```python
import string
string.punctuation
```

```python
def remove_punct(text):
  remove_pun="".join([word for word in text if word not in string.punctuation])
  return remove_pun

books['text_without_pun']=books['text'].apply(lambda x:remove_punct(x))
books.head(5)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwords_Ar=nltk.corpus.stopwords.words('arabic')
stopwords_Ar
```

```python
from nltk import word_tokenize
nltk.download('punkt')
def remove_stopwords (text):
  wordss=word_tokenize(text)
  filtered_words=[word for word in wordss if word not in stopwords_Ar]
  filtered_text=" ".join(filtered_words)
  return filtered_text
books['text_without_stopwords']=books['text_without_pun'].apply(lambda x:remove_stopwords(x))
books.head(10)
```

```python
def tokenize(text):
  token=word_tokenize(text)
  return token
books['text_tokenize']=books['text_without_stopwords'].apply(lambda x: tokenize(x))
books.head(5)
```

```python
from snowballstemmer import stemmer
ar_stemmer= stemmer("arabic")
def stem(s):
  w=word_tokenize(s)
  words=[ar_stemmer.stemWord(word) for word in w]
  text=" ".join(words)
  return text
books['stemming']=books['text_without_stopwords'].apply(lambda x: stem(x))
books.head(5)

```

```python
from nltk import pos_tag
from nltk import word_tokenize
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
def pos_tok(text):
  return pos_tag(text)
books['POS']=books['text_tokenize'].apply(lambda x: pos_tok(x))
books.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(ngram_range=(1,3))
features_cv=vectorizer.fit_transform(books['stemming'])
print(features_cv.shape)
print(features_cv)
```

```python
re=books['label']
re
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder =LabelEncoder()
y_encoder=label_encoder.fit_transform(re)
```

```python
from sklearn.model_selection import train_test_split
X_tarin, X_test, y_train, y_test = train_test_split(features_cv,y_encoder,test_size=0.25, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_tarin, y_train)
```

```python
accuracy=clf.score(X_test, y_test)
print(f"accuracy={accuracy:.2f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
clf2= RandomForestClassifier(n_jobs=-1)
clf2.fit(X_tarin,y_train)
```

```python
accuracy=clf2.score(X_test, y_test)
print(f"accuracy={accuracy:.2f}")
```
